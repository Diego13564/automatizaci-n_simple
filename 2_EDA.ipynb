```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('./data/titanic_clean.csv')

print("¡Librerías importadas y datos cargados con éxito!")
```

```python

plt.figure(figsize = (12, 8))

sns.heatmap(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(), 
            annot=True, 
            cbar=True,  
            cmap='coolwarm',  
            cbar_kws={'label': 'Nivel de correlación'})  

plt.title('Correlación de variables numéricas')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', kde=True, color='skyblue')
plt.title('Distribución de Edades')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Pclass', hue='Survived', palette=['#8B0000', '#006400'])
plt.title('Sobrevivientes vs Fallecidos por Clase')
plt.legend(title='Supervivencia', labels=['Falleció', 'Sobrevivió'])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Sex', hue='Survived', palette='Set2')
plt.title('Sobrevivientes por Género')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Embarked', hue='Survived', palette='viridis')
plt.title('Sobrevivientes por Puerto de Embarque')
plt.show()
```

```python
# Primero creamos los grupos y las etiquetas
bins = [0, 5, 12, 18, 35, 60, 100]
labels = ['Infante', 'Niño', 'Adolescente', 'Joven Adulto', 'Adulto', 'Mayor']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

plt.figure(figsize=(10, 6))
# Usamos barplot porque automáticamente calcula el promedio (tasa) de supervivencia
sns.barplot(data=df, x='AgeGroup', y='Survived', palette='magma')
plt.title('Tasa de Supervivencia por Grupo de Edad')
plt.ylabel('Tasa de Supervivencia (0 a 1)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Pclass', palette='Blues_d')
plt.title('Total de pasajeros por Clase')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Pclass', hue='Sex', palette='Pastel1')
plt.title('Hombres y Mujeres por Clase')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Embarked', hue='Pclass', palette='Set3')
plt.title('Clases de pasajeros por Puerto de Embarque')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sex', palette=['#1f77b4', '#ff7f0e'])
plt.title('Total de Hombres y Mujeres')
plt.show()
```

```python
print(f"El pasajero más joven tenía: {df['Age'].min()} años.")
print(f"El pasajero más grande tenía: {df['Age'].max()} años.")
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(data=df, x='Age', kde=True, color='lightgray')
plt.axvline(df['Age'].mean(), color='red', label=f"Media: {df['Age'].mean():.1f}")
plt.axvline(df['Age'].median(), color='green', label=f"Mediana: {df['Age'].median():.1f}")
plt.axvline(df['Age'].mode()[0], color='black', label=f"Moda: {df['Age'].mode()[0]:.1f}")
plt.title('Distribución de Edades con Medidas Centrales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Age', color='lightgreen')
plt.title('Boxplot de Edades (Detección de Outliers)')
plt.show()
```

```python
print(f"La tarifa más barata fue: ${df['Fare'].min():.2f}")
print(f"La tarifa más cara fue: ${df['Fare'].max():.2f}")
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(data=df, x='Fare', kde=True, bins=40, color='lightblue')
plt.axvline(df['Fare'].mean(), color='red', label=f"Media: {df['Fare'].mean():.2f}")
plt.axvline(df['Fare'].median(), color='green', label=f"Mediana: {df['Fare'].median():.2f}")
plt.axvline(df['Fare'].mode()[0], color='black', label=f"Moda: {df['Fare'].mode()[0]:.2f}")
plt.title('Distribución de Tarifas')
plt.xlim(0, 200) # Limitamos la vista para que los valores extremos no aplasten la gráfica
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x='Fare', color='salmon')
plt.title('Boxplot de Tarifas (Outliers)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='SibSp', palette='rocket')
plt.title('Pasajeros por cantidad de Hermanos/Esposos a bordo')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='SibSp', y='Survived', palette='mako')
plt.title('Tasa de Supervivencia según Hermanos/Esposos')
plt.ylabel('Tasa de supervivencia')
plt.show()
```

```python
porcentajes_sibsp = df.groupby('SibSp')['Survived'].mean() * 100
print("Porcentaje de supervivencia por número de hermanos/esposos:")
print(porcentajes_sibsp.round(2).astype(str) + ' %')
```

```python
plt.figure(figsize=(10, 8))
# Seleccionamos solo las columnas numéricas para evitar errores matemáticos
columnas_numericas = df.select_dtypes(include=['number'])

sns.heatmap(columnas_numericas.corr(), annot=True, cmap='RdYlGn', linewidths=0.5)
plt.title('Matriz de Correlación Final')
plt.show()
```
