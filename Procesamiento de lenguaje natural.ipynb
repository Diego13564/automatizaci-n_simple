```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# 1. Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(train_x_vector, train_y)

acc_log = log_reg.score(test_x_vector, test_y)
print(f"Precisión Logistic Regression: {acc_log:.4f}")
```

```python
# 2. Gaussian Naive Bayes
gnb = GaussianNB()
# Usamos .toarray() para que el modelo lo entienda
gnb.fit(train_x_vector.toarray(), train_y)

acc_gnb = gnb.score(test_x_vector.toarray(), test_y)
print(f"Precisión GaussianNB: {acc_gnb:.4f}")
```

```python
# 3. Decision Tree
tree = DecisionTreeClassifier(random_state=42)
tree.fit(train_x_vector, train_y)

acc_tree = tree.score(test_x_vector, test_y)
print(f"Precisión Decision Tree: {acc_tree:.4f}")
```
