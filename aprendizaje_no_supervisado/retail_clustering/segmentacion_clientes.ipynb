```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

# Carga de datos (Asegúrate de que el archivo esté en la carpeta /data)
df = pd.read_csv('./data/retailmax.csv')
print("¡Datos cargados! Aquí tienes las primeras filas:")
df.head()
```

```python
# Seleccionamos las columnas numéricas
features = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Escalamos los datos
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

print("Datos escalados y listos para procesar.")
```

```python
wcss = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', color='red')
plt.title('Buscando el número óptimo de grupos (Codo)')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('WCSS (Inercia)')
plt.show()
```

```python
# Aplicamos K-Means con 5 clusters
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

print("¡Hecho! Hemos asignado un grupo a cada cliente.")
df.head()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Dibujamos los puntos
scatter = ax.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], df['Age'],
                     c=df['Cluster'], cmap='Set1', s=50)

ax.set_xlabel('Ingreso Anual')
ax.set_ylabel('Puntaje de Gasto')
ax.set_zlabel('Edad')
plt.title('Visualización 3D de Segmentos de Clientes')
plt.show()
```
